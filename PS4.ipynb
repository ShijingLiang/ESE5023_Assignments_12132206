## Global plot

```python
df_fil=df.iloc[:,1:5].dropna()
len(df),len(df_fil)
```

```python
eq_max = xr.DataArray(
    np.zeros((71,143)),
    [('lat',np.arange(-87.5,88,2.5)),
     ('lon',np.arange(-177.5,180,2.5))])

dep_max = xr.DataArray(
    np.zeros((71,143)),
    [('lat',np.arange(-87.5,88,2.5)),
     ('lon',np.arange(-177.5,180,2.5))])

eq_mean = xr.DataArray(
    np.zeros((71,143)),
    [('lat',np.arange(-87.5,88,2.5)),
     ('lon',np.arange(-177.5,180,2.5))])

dep_mean = xr.DataArray(
    np.zeros((71,143)),
    [('lat',np.arange(-87.5,88,2.5)),
     ('lon',np.arange(-177.5,180,2.5))])

eq_count = xr.DataArray(
    np.zeros((71,143)),
    [('lat',np.arange(-87.5,88,2.5)),
     ('lon',np.arange(-177.5,180,2.5))])
```

```python
lat=np.arange(-87.5,88,2.5)
lon=np.arange(-177.5,180,2.5)
```

```python
for x in range(71):
    for y in range(143):
            data=df_fil.loc[(df_fil['latitude']>=lat[x]-1.25)
                               & (df_fil['latitude']<=lat[x]+1.25) 
                               & (df_fil['longitude']>=lon[y]-1.25)
                               & (df_fil['longitude']<=lon[y]+1.25)]
            
            eq_max[x][y]=data['mag'].max()
            dep_max[x][y]=data['depth'].max()
            eq_mean[x][y]=data['mag'].mean()
            dep_mean[x][y]=data['depth'].mean()
            eq_count[x][y]=len(data)
            
    if (x%20==0):        
        print(x)
        
'''
eq_max.to_netcdf('2014_global_earthquakes_max_mag.nc')
dep_max.to_netcdf('2014_global_earthquakes_max_depth.nc')
eq_mean.to_netcdf('2014_global_earthquakes_mean_mag.nc')
dep_mean.to_netcdf('2014_global_earthquakes_mean_depth.nc')
eq_count.to_netcdf('2014_global_earthquakes_count.nc')
'''    
```

```python
eq_max=xr.open_dataset('2014_global_earthquakes_max_mag.nc')['__xarray_dataarray_variable__']
dep_max=xr.open_dataset('2014_global_earthquakes_max_depth.nc')['__xarray_dataarray_variable__']
eq_mean=xr.open_dataset('2014_global_earthquakes_mean_mag.nc')['__xarray_dataarray_variable__']
dep_mean=xr.open_dataset('2014_global_earthquakes_mean_depth.nc')['__xarray_dataarray_variable__']
eq_count=xr.open_dataset('2014_global_earthquakes_count.nc')['__xarray_dataarray_variable__']
```

```python
proj = ccrs.Robinson(central_longitude=180)
fig,axs=plt.subplots(2,2,subplot_kw=dict(projection=proj),figsize=(20,20))
plt.subplots_adjust(hspace=-0.5) 

for axi in axs:
    for ax in axi:
        ax.set_global()
        ax.stock_img()
        ax.gridlines(draw_labels=True)
        
font={'size':20}
x,y=np.meshgrid(lon,lat)

sc1=axs[0,0].scatter(x,y,c=eq_max,zorder=4,cmap='hot_r',alpha=0.7,lw=.5,transform=ccrs.PlateCarree())

cb1=plt.colorbar(sc1,ax=axs[0,0],shrink=.4,pad=.03,extend='both',location='bottom')
cb1.set_label('Max magnitude',fontdict=font)

sc2=axs[0,1].scatter(x,y,c=dep_max,zorder=4,cmap='plasma',alpha=0.7,lw=.5,transform=ccrs.PlateCarree())
cb2=plt.colorbar(sc2,ax=axs[0,1],shrink=.4,pad=.03,extend='both',location='bottom')
cb2.set_label('Max depth',fontdict=font)

sc3=axs[1,0].scatter(x,y,c=eq_mean,zorder=4,cmap='hot_r',alpha=0.7,lw=.5,transform=ccrs.PlateCarree())
cb3=plt.colorbar(sc3,ax=axs[1,0],shrink=.4,pad=.03,extend='both',location='bottom')
cb3.set_label('Mean magnitude',fontdict=font)

sc4=axs[1,1].scatter(x,y,c=dep_mean,zorder=4,cmap='plasma',alpha=0.7,lw=.5,transform=ccrs.PlateCarree())
cb4=plt.colorbar(sc4,ax=axs[1,1],shrink=.4,pad=.03,extend='both',location='bottom')
cb4.set_label('Mean depth',fontdict=font)

axs[0,0].set_title('Global earthquakes 2014',fontsize=26,loc='left')
```

## Regional plot

```python
def plot_map():
    proj = ccrs.Robinson(central_longitude=180)
    fig,ax = plt.subplots(figsize=(16,16),subplot_kw=dict(projection=proj))
    ax.set_global()
    ax.stock_img()
    ax.coastlines()
    ax.gridlines(draw_labels=True)
    plt.tick_params(axis='both',labelsize=30)
    return fig,ax
```

```python
fig,ax = plot_map()
ax.set_extent([180,300,10,60])
x,y=np.meshgrid(lon,lat)
scatter=ax.scatter(x,y,s=eq_count*.5,c=eq_count,zorder=4,cmap='Reds',edgecolors='k',lw=.5,alpha=0.7,transform=ccrs.PlateCarree())

font={'size':20}
plt.colorbar(scatter,shrink=.4,pad=.1,extend='both',label='Count')

plt.title('North America earthquakes counts at 2014',fontsize=20)

# produce a legend with a cross section of sizes from the scatter
#handles, labels = scatter.legend_elements(prop="sizes", alpha=0.6)
legend = ax.legend(*scatter.legend_elements(prop='sizes',num=6,
                                           ), loc="upper right", title="Sizes",markerscale=.5,
                   fontsize=15,scatteryoffsets=[0,.25,.75,1],title_fontsize=15)
```
